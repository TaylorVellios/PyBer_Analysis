```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statistics
```

```python
file = 'Resources/PyBer_ride_data.csv'
openfile = pd.read_csv(file)

ride_data_df = pd.DataFrame(openfile)

ride_data_df
```

# USINIG .PLOT() DIRECTLY WITH A DATAFRAME

```python
ride_data_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
#SEE HOW WE AREN'T GETTING ALL OF THE TICKS? LET'S FIX THAT.
x_axis = np.arange(len(ride_data_df))

tick_locs = [i for i in x_axis]

ride_data_df.plot(x="Month", y="Avg. Fare ($USD)", figsize=(10,4))
plt.xticks(tick_locs, ride_data_df["Month"])
plt.show()
```

```python
ride_data_df.plot(x="Month", y="Avg. Fare ($USD)", kind="bar")
plt.show()
```

```python
#SKILL DRILL 5.1.10
y_axis = [i for i in ride_data_df["Avg. Fare ($USD)"]]
stdev = statistics.stdev(y_axis)

ride_data_df.plot(x="Month", y="Avg. Fare ($USD)", kind="bar", color="skyblue", yerr=stdev, capsize=5, figsize=(10,6))
plt.xticks(rotation=0)
plt.yticks([i for i in range(51) if i==0 or i%5==0])

plt.show()
```
